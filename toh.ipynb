# Tower Of Hanoi (Recursive)

```python
#function to print path

def printMoves(n, start, end):
    print("Move disk {} from {} to {}".format(n, start, end))
```

```python
#recursive toh function that prints disk movement

def toh(rods, src, dest, aux):
    if rods==1:
        print("Move disk 1 from {} to {}".format(src, dest))
    else:
        toh(rods-1, src, aux, dest)
        printMoves(rods, src, dest)
        toh(rods-1, aux, dest, src)
```

```python
#recursive toh function to count disk movement (does not print movement)

def tohCountMoves(rods, src, dest, aux):
    moves = 1
    if rods==1:
        return moves
    else:
        moves += tohCountMoves(rods-1, src, aux, dest)
        moves += tohCountMoves(rods-1, aux, dest, src)
    return moves
```

```python
#print disk movement for upto 5 disks

for x in range(1,6):
    print("For {} disk/disks, the output is as follows: \n".format(x))
    toh(x, 'A', 'B', 'C')
    print("---------------------------------------------------------------------------")
```

```python
import timeit
import pandas as pd
```

```python
#get total moves and total time taken by toh function for upto 15 disks

d = []
for x in range(1,16):
    nMove = tohCountMoves(x, 'A','B','C')
    time = timeit.Timer(lambda: tohCountMoves(x, 'A','B','C'))
    calc = (2**x)-1
    d.append([x, nMove, calc, time.timeit(5)])

data = pd.DataFrame(d, columns=['No. of disks','No. of moves','2^N-1','Time Taken'])
data
```
